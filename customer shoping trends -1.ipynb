```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder



# Selecting relevant numerical features for clustering
clustering_features = df[[ 'Purchase Amount (USD)', 'Previous Purchases']]



```

```python
from sklearn.preprocessing import LabelEncoder

# Creating a copy of the dataframe to apply label encoding
df_encoded = df.copy()

# List of categorical columns to encode
categorical_cols = ['Age','Gender', 'Item Purchased', 'Category','Review Rating', 'Location', 'Size', 'Color', 
                    'Season', 'Subscription Status', 'Payment Method', 'Shipping Type', 
                    'Discount Applied', 'Promo Code Used', 'Preferred Payment Method', 
                    'Frequency of Purchases']

# Applying label encoding to each categorical column
label_encoder = LabelEncoder()
for col in categorical_cols:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])

# Display the first few rows of the encoded dataframe
df_encoded=df_encoded.drop("Age Group", axis=1)
df_encoded.head()
```

```python
# Combining the encoded categorical features with the numerical features
combined_features = pd.concat([clustering_features, df_encoded], axis=1)
combined_features=combined_features.drop("Customer ID", axis=1)
combined_features.head()

```

```python
# Standardizing the combined features
scaler = StandardScaler()
scaled_combined_features = scaler.fit_transform(combined_features)

# Re-applying the elbow method with the combined features
wcss_combined = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_combined_features)
    wcss_combined.append(kmeans.inertia_)

# Plotting the results of the elbow method with combined features
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss_combined)
plt.title('Elbow Method For Optimal k (with Categorical Features)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting k-means using 3 clusters
kmeans_3 = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans_3.fit(scaled_combined_features)

# Fitting k-means using 4 clusters
kmeans_4 = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans_4.fit(scaled_combined_features)

# Adding the cluster labels to the original dataframe for interpretation
df['Cluster_Labels_3'] = kmeans_3.labels_
df['Cluster_Labels_4'] = kmeans_4.labels_

# We can now look at the cluster centers or the count of items in each cluster for further analysis.
cluster_centers_3 = kmeans_3.cluster_centers_
cluster_centers_4 = kmeans_4.cluster_centers_

cluster_centers_3, cluster_centers_4, df['Cluster_Labels_3'].value_counts(), df['Cluster_Labels_4'].value_counts()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# We will use PCA to reduce the data to two dimensions for visualization
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_combined_features)

# Scatter plot for 3 clusters
plt.figure(figsize=(14, 7))

# Subplot for 3 clusters
plt.subplot(1, 2, 1)
plt.scatter(reduced_data[df['Cluster_Labels_3'] == 0, 0], reduced_data[df['Cluster_Labels_3'] == 0, 1], s=10, c='red', label='Cluster 1')
plt.scatter(reduced_data[df['Cluster_Labels_3'] == 1, 0], reduced_data[df['Cluster_Labels_3'] == 1, 1], s=10, c='blue', label='Cluster 2')
plt.scatter(reduced_data[df['Cluster_Labels_3'] == 2, 0], reduced_data[df['Cluster_Labels_3'] == 2, 1], s=10, c='green', label='Cluster 3')
plt.title('Visualization of Clusters (k=3)')
plt.legend()

# Subplot for 4 clusters
plt.subplot(1, 2, 2)
plt.scatter(reduced_data[df['Cluster_Labels_4'] == 0, 0], reduced_data[df['Cluster_Labels_4'] == 0, 1], s=10, c='red', label='Cluster 1')
plt.scatter(reduced_data[df['Cluster_Labels_4'] == 1, 0], reduced_data[df['Cluster_Labels_4'] == 1, 1], s=10, c='blue', label='Cluster 2')
plt.scatter(reduced_data[df['Cluster_Labels_4'] == 2, 0], reduced_data[df['Cluster_Labels_4'] == 2, 1], s=10, c='green', label='Cluster 3')
plt.scatter(reduced_data[df['Cluster_Labels_4'] == 3, 0], reduced_data[df['Cluster_Labels_4'] == 3, 1], s=10, c='purple', label='Cluster 4')
plt.title('Visualization of Clusters (k=4)')
plt.legend()

plt.show()
```
